```python
# NOW LETS LOOK AT THE CONCEPT OF CREATING RANDOM TENSORS 
# random tensors are tensors of some arbitrary size which contains random numbers 
# random tensors are used by nueral networks to initialise weights(patterns) that are trying to learn data 
# 1. INITIALISE WITH RANDOM WEIGHTS(BEGINNING) -> SHOW EXAMPLES TO MODEL -> UPDATE -> REPEAT WITH MORE EXAMPLES 
# INPUTS -> NUMERICAL ENCODING -> LEARN REPRESENTATION(PATTERNS/FEATURES/WEIGHTS) -> OUTPUTS  
```

```python
# CREATING A RANDOM TENSOR 
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility 

random_1 = random_1.normal(shape=(3, 2)) # create tensor from normal distribution 

random_2 = tf.random.Generator.from_seed(42) 

random_2 = random_2.normal(shape=(3, 2)) 
```

```python
# check if there are equal ? yes if a seed is set they will be equal 
random_1, random_2
```

```python
# NOW WHAT WILL HAPPEN IS THE SEED IS CHANGED 
# CREATE A RANDOM SEED OF 42 AND 11 THAT TAKES THE SHAPE OF A 3, 2 
random_3 = tf.random.Generator.from_seed(42) 
random_3 = random_3.normal(shape=(3, 2))

random_4 = tf.random.Generator.from_seed(11) 
random_4 = random_4.normal(shape=(3, 2))

# now check for equality 
random_3, random_4, random_1 

# random_3 == random_3
random_3 == random_1
random_4 == random_3
```

```python
# Create (3, 2) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])

# Create another (3, 2) tensor
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])
X, Y
```

```python
X @ Y 
```

```python
# either reshape X to (2, 3) so its (2, 3) @ (3, 2) 
# reshape Y to (3, 2) so its (3, 2) @ (2, 3) 
# i can use either tf.transpose() => switches the dimensions of a given tensor or tf.reshape() => allows us to reshape into a defined shape 
```

```python
tf.reshape(Y, shape=(2, 3))
```

```python
# Try matrix multipilication with reshaped Y 
X @ tf.reshape(Y, shape=(2, 3))
```

```python
# use transpose 
tf.transpose(X) 
```

```python
# Try matrix multiplication 
tf.matmul(tf.transpose(X), Y)
```

```python
# You can achieve the same result with parameters
tf.matmul(a=X, b=Y, transpose_a=True, transpose_b=False)
```

```python
# (3, 2) @ (2, 3) -> (3, 3) done with X @ tf.reshape(Y, shape=(2, 3)) 
# (2, 3) @ (3, 2) -> (2, 2) done with tf.matmul(tf.transpose(X), Y)
```

```python
# This kind of data manipulation is a reminder: you'll spend a lot of your time in machine learning and working with 
# neural networks reshaping data (in the form of tensors) to prepare it to be used with various operations (such as feeding it to a model).
```

```python
# DOT PRODUCT => CAN PERFORM THE TF.MATMUL() OPERATION USING TF.TENSORDOT() 
tf.tensordot(tf.transpose(X), Y, axes=1) 
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, (2, 3)))
```

```python
# WE GET DIFFERENT VALUES 
# THIS IS STRANGE BCOS DEALING WITH Y a (3x2) matrix reshaping to (2, 3) and transposing it result in the same shape 
# Check shapes of Y, reshaped Y and Y and transposed Y 
Y.shape, tf.reshape(Y, (2, 3)).shape, tf.transpose(Y).shape 
```

```python
# CALLING TF.RESHAPE() AND TF.TRANSPOSE() ON Y dont result in the same values 
print("Normal Y:")
print(Y, "\n")
```

```python
print("Y reshaped to (2, 3): ")
print(tf.reshape(Y, (2, 3)), "\n")
```

```python
print("Y transposed: ")
print(tf.transpose(Y))
```

```python
# SOME EXPLANATIONS 
# tf.reshape() - change the shape of the given tensor (first) and then insert values in order they appear (in our case, 7, 8, 9, 10, 11, 12)
# tf.transpose() - swap the order of the axes, by default the last axis becomes the first, however the order can be changed using the perm parameter.
# But generally, whenever performing a matrix multiplication and the shapes of two matrices don't line up, you will transpose (not reshape) one of them in order to line them up.
```

```python
# USING TF.FUNCTION 
# In your TensorFlow adventures, you might come across Python functions which have the decorator @tf.function
# Its a decorator that modifies functions in another way 
# In the @tf.function decorator case, it turns a Python function into a callable TensorFlow graph. 
# Which is a fancy way of saying, if you've written your own Python function, and you decorate it with @tf.function, when you export your code (to potentially run on another device), 
# TensorFlow will attempt to convert it into a fast(er) version of itself (by making it part of a computation graph).

```

```python
# create a simple function 
def function(x, y): 
    return x * y 

x = tf.constant(np.arange(0, 10))

y = tf.constant(np.arange(10, 20))

# function(x, y) 
```

```python
x 
```

```python
y 
```

```python
function(x, y)
```

```python
# CREATE THE SAME FUNCTION AND DECORATE IT WITH TF.FUNCTION 
@tf.function 
def tf_function(x, y):
    return x * y 
tf_function(x, y) 
```
