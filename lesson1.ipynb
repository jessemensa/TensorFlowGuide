```python
# NOW LETS LOOK AT THE CONCEPT OF CREATING RANDOM TENSORS 
# random tensors are tensors of some arbitrary size which contains random numbers 
# random tensors are used by nueral networks to initialise weights(patterns) that are trying to learn data 
# 1. INITIALISE WITH RANDOM WEIGHTS(BEGINNING) -> SHOW EXAMPLES TO MODEL -> UPDATE -> REPEAT WITH MORE EXAMPLES 
# INPUTS -> NUMERICAL ENCODING -> LEARN REPRESENTATION(PATTERNS/FEATURES/WEIGHTS) -> OUTPUTS  
```

```python
# CREATING A RANDOM TENSOR 
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility 

random_1 = random_1.normal(shape=(3, 2)) # create tensor from normal distribution 

random_2 = tf.random.Generator.from_seed(42) 

random_2 = random_2.normal(shape=(3, 2)) 
```

```python
# check if there are equal ? yes if a seed is set they will be equal 
random_1, random_2
```

```python
# NOW WHAT WILL HAPPEN IS THE SEED IS CHANGED 
# CREATE A RANDOM SEED OF 42 AND 11 THAT TAKES THE SHAPE OF A 3, 2 
random_3 = tf.random.Generator.from_seed(42) 
random_3 = random_3.normal(shape=(3, 2))

random_4 = tf.random.Generator.from_seed(11) 
random_4 = random_4.normal(shape=(3, 2))

# now check for equality 
random_3, random_4, random_1 

# random_3 == random_3
random_3 == random_1
random_4 == random_3
```

```python
# MAIN GOAL OF THIS IS LETS SAY YOU ARE WORKING WITH 15,000 IMAGES OF CATS AND DOGS AND THE FIRST 10,000 IMAGES 
# ARE CATS AND 5,000 IMAGES ARE DOGS 
# IT IS A GOOD IDEA TO MOVE THE DATA AROUND AS THIS WILL AFFECT HOW A NN LEARNS 
```

```python
# SHUFFLE THE TENSOR 
not_shuffled = tf.constant([[10, 7], [3, 4], [2, 5]])
# gets different results each time 
tf.random.shuffle(not_shuffled)
```

```python
# why is the value above and below different ?? 
tf.random.shuffle(not_shuffled, seed=42)
```

```python
# set a global random seed 
tf.random.set_seed(42) 

# set the operation random seed 
tf.random.shuffle(not_shuffled, seed=42) 
```

```python
# Set the global random seed
tf.random.set_seed(42) # if you comment this out you'll get different results

# Set the operation random seed
tf.random.shuffle(not_shuffled)
```

```python
# MAIN DIFFERENCE BETWEEN TENSORS AND NUMPY ARRAYS IS THAT TENSORS CAN RUN ON GPUs 

# GETTING INFORMATION FROM TENSORS(SHAPE, RANK AND SIZE) 
# SHAPE => THE LENGTH(NUMBER OF ELEMENT OF EACH OF THE DIMENSIONS OF A TENSOR)
# RANK => THE NUMBER OF TENSOR DIMENSIONS. A SCALAR HAS A RANK 0, VECTOR HAS A RANK 1, MATRIX RANK 2, TENSOR RANK N 
# AXIS OR DIMENSIONS => A PARTICULAR DIMENSION OF A TENSOR 
# SIZE => THE TOTAL NUMBER OF ITEMS IN THE TENSOR 
```

```python
# CREATE A RANK 4 TENSOR 
rank4 = tf.zeros([2, 3, 4, 5, 6])
rank4 
```

```python
rank4.shape, rank4.ndim, tf.size(rank4)
```

```python
# GET THE VARIOUS ATTRIBUTES OF THE TENSOR 
print("Datatype of every element:", rank4.dtype)
print("Number of dimensions (rank):", rank4.ndim)
print("Shape of tensor:", rank4.shape)
print("Elements along axis 0 of tensor:", rank4.shape[0])
print("Elements along last axis of tensor:", rank4.shape[-1])
print("Total number of elements (2*3*4*5):", tf.size(rank4).numpy()) # .numpy() converts to NumPy array
```

```python
# GET THE FIRST 2 ITEMS OF EACH DIMENSIONS 
rank4[:2, :2, :2, :2]
```

```python
# GET THE DIMENSION FROM EACH INDEX EXCEPT FOR THE FINAL ONE 
rank4[:1, :1, :1, :]
```

```python
# CREATE A RANK 2 TENSOR (2 DIMENSIONS) 
rank2 = tf.constant([[10, 7], [3, 4]])

# get the last item of each row 
rank2[:, -1]
```

```python
# ADD AN EXTRA DIMENSION (TO THE END) 
rank3 = rank2[..., tf.newaxis] # ... this means to all dimensions prior to 
rank2, rank3 # 
```
